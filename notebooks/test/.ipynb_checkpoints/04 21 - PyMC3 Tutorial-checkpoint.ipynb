```python
import pymc3 as pm

import pandas as pd
import janitor

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 100000

# .5 percent of the population are users
users = np.random.binomial(n, 0.005, 1)
non_users = n - users
```

```python
# True positive rate is 99%
user_positives = np.random.binomial(users, 0.99)
non_user_positives = np.random.binomial(non_users, 0.01)
```

```python
# Fraction of positives that are users
user_positives / (user_positives + non_user_positives)
```

# Click-through rate analysis

```python
ctr = pd.read_csv(
    'https://raw.githubusercontent.com/ericmjl/bayesian-stats-modelling-tutorial/master/data/clickthrough.csv',
    index_col=0
).label_encode('group')

ctr.sample(10)
```

```python
ctr.groupby('group').mean()
```

```python
ctr.groupby('group').count()
```

```python
control_df = ctr.query('group == "control"')

with pm.Model() as model1_bernoulli:
    p = pm.Uniform('p', lower=0, upper=1)
    like = pm.Bernoulli('like', p=p, observed=control_df['clicks'])
```

```python
with model1_bernoulli:
    samples = pm.sample(2000)
```

```python
pm.plot_posterior(samples, kind='hist')
plt.show()
```

```python
pm.plot_posterior(samples)
plt.show()
```

```python
test_df = ctr.query('group == "test"')

with pm.Model() as model2_bernoulli:
    # Control group
    p_control = pm.Uniform('p_control', lower=0, upper=1)
    like_control = pm.Bernoulli(
        'like_control', 
        p=p_control, 
        observed=control_df['clicks']
    )
    
    # Test group
    p_test = pm.Uniform('p_test', lower=0, upper=1)
    like_test = pm.Bernoulli(
        'like_test',
        p=p_test,
        observed=test_df['clicks']
    )
    
    # Difference of the p's
    # Math on probability distributions
    p_diff = pm.Deterministic('p_diff', p_test - p_control)
```

```python
with model2_bernoulli:
    samples = pm.sample(2000)
```

```python
pm.plot_posterior(samples)
plt.show()
```

```python
samples['p_diff']
```

```python
dollar_distr = samples['p_diff'] * 25 * 1e6
n = len(dollar_distr)

plt.plot(
    np.sort(dollar_distr), 
    np.arange(1, n + 1) / n
)

plt.xlabel('Expected revenue increase')
plt.ylabel('Cumulative probability')

plt.show()
```

# Baseball data

```python
players = (
    pd.read_csv('https://raw.githubusercontent.com/ericmjl/bayesian-stats-modelling-tutorial/master/data/baseballdb/core/Batting.csv')
    .clean_names()  # FYI: pyjanitor
    .query("yearid == 2016")
    .select_columns(['playerid', 'ab', 'h'])  # FYI: pyjanitor
    .groupby('playerid').sum()
)

salaries = (
    pd.read_csv('https://raw.githubusercontent.com/ericmjl/bayesian-stats-modelling-tutorial/master/data/baseballdb/core/Salaries.csv')
    .clean_names()  # FYI: pyjanitor
    .query("yearid == 2016")
    .select_columns(["playerid", "salary"])  # FYI: pyjanitor
    .groupby('playerid').mean()
)

data = (
    players.join(salaries)
    .dropna()
    .reset_index().label_encode("playerid").set_index("playerid")
)
data
```

```python
with pm.Model() as pitch_model:
    p = pm.Beta('p', alpha=1, beta=1, shape=(len(data)))
    like = pm.Binomial('like', p=p, n=data['ab'], observed=data['h'])
    
    # p per salary
    pps = pm.Deterministic('pps', p / data['salary'])
```

```python
with pitch_model:
    trace_pitch = pm.sample(2000)
```

```python
from ipywidgets import SelectMultiple, HBox, VBox, Select, fixed, interact

def dict2tuple(d):
    return list(*zip(*zip(d.items())))

def inversedict(d):
    return {v:k for k, v in d.items()}

playerid_mapping = data['playerid_enc'].to_dict()
playerid_inverse = inversedict(playerid_mapping)

player_select = SelectMultiple(
    options=dict2tuple(playerid_mapping),
    value=(0,)
)

metric_select = Select(
    options=['p', 'pps'],
)
```

```python
def scrub_posterior(player_encs: list, metric: str, trace):
    fig, ax = plt.subplots()
    
    for enc in player_encs:
        data = trace[metric][:, enc]
        n = len(data)

        ax.plot(
            np.sort(data), 
            np.arange(1, n + 1) / n,
            label=playerid_inverse[enc]
        )
        
    ax.legend()
    ax.set_xlabel("p per salary")
    ax.set_ylabel('cumulative probability')
        
    sns.despine()
    
    
interact(scrub_posterior, player_encs=player_select, metric=metric_select, trace=fixed(trace_pitch));
```

```python
with pm.Model() as hier_baseball:
    a_prior = pm.Exponential('a_prior', lam=1 / 29.)
    b_prior = pm.Exponential('b_prior', lam=1 / 29.)
    
    p = pm.Beta('p', alpha=a_prior, beta=b_prior, shape=len(data))
    like = pm.Binomial('like', p=p, n=data['ab'], observed=data['h'])
    
    pps = pm.Deterministic('pps', p / data['salary'])
```

```python
with hier_baseball:
    trace_hier = pm.sample(2000)
```

```python
pm.plot_posterior(trace_hier)
plt.show()
```
