```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')

import sklearn
```

```python
types = ['ES', 'PSH', 'TH', 'RRH', 'PREV']

df = pd.read_csv('../../data/subset_data.csv', index_col=0)
# df = df[df['Real'] != 2]
df.head()
```

```python
mean_prob_df = df.sort_values('Week').groupby('Week').mean()[types]

print(mean_prob_df.shape)
mean_prob_df.head()
```

# Begin example from Scikit-learn documentation

```python
from sklearn.datasets import fetch_openml

data = fetch_openml(data_id=41187)

months = []
ppmv_sums = []
counts = []

y = data.data[:, 0]
m = data.data[:, 1]
month_float = y + (m - 1) / 12
ppmvs = data.target

for month, ppmv in zip(month_float, ppmvs):
    if not months or month != months[-1]:
        months.append(month)
        ppmv_sums.append(ppmv)
        counts.append(1)
    else:
        ppmv_sums[-1] += ppmv
        counts[-1] += 1
    
months = np.array(months).reshape(-1, 1)
avg_ppmvs = np.array(ppmv_sums) / counts
    
X, y = months, avg_ppmvs

print(X)
print('=' * 40)
print(y)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, WhiteKernel, RationalQuadratic, ExpSineSquared
```

```python
# Kernels with parameters given in the GPML book
k1 = 66.0 ** 2 * RBF(length_scale=67.0)  # long term smooth rising trend
k2 = 2.4 ** 2 * RBF(length_scale=90.0) \
     * ExpSineSquared(length_scale=1.3, periodicity=1.0)  # seasonal component

# Medium term irregularity
k3 = 0.66 ** 2 * RationalQuadratic(length_scale=1.2, alpha=0.78)
k4 = 0.18 ** 2 * RBF(length_scale=0.134) + WhiteKernel(noise_level=0.19 ** 2)  # noise terms

kernel_gpml = k1 + k2 + k3 + k4

gp = GaussianProcessRegressor(kernel=kernel_gpml, alpha=0, optimizer=None, normalize_y=True)
gp.fit(X, y)

print(f'GPML kernel: {gp.kernel_}')
print(f'Log marginal likelihood: {gp.log_marginal_likelihood(gp.kernel_.theta):.3f}')
```

```python
plt.figure(figsize=(20, 15))

X_ = np.linspace(X.min(), X.max() + 30, 1000)[:, np.newaxis]
y_pred, y_std = gp.predict(X_, return_std=True)

plt.scatter(X, y, c='k')
plt.plot(X_, y_pred)
plt.fill_between(X_[:, 0], y_pred - y_std, y_pred + y_std, alpha=0.5, color='k')
plt.xlim(X_.min(), X_.max())

plt.xlabel('Year')
plt.ylabel('CO2 in ppm')
plt.tight_layout()
plt.show()
```

```python
# Kernels with parameters given in the GPML book
k1 = 66.0 ** 2 * RBF(length_scale=67.0)  # long term smooth rising trend
k2 = 2.4 ** 2 * RBF(length_scale=90.0) \
     * ExpSineSquared(length_scale=1.3, periodicity=1.0)  # seasonal component

# Medium term irregularity
k3 = 0.66 ** 2 * RationalQuadratic(length_scale=1.2, alpha=0.78)
k4 = 0.18 ** 2 * RBF(length_scale=0.134) + WhiteKernel(noise_level=0.19 ** 2)  # noise terms

kernel = k1 + k2 + k3 + k4

gp = GaussianProcessRegressor(kernel=kernel, alpha=0, normalize_y=True)
gp.fit(X, y)

print(f'GPML kernel: {gp.kernel_}')
print(f'Log marginal likelihood: {gp.log_marginal_likelihood(gp.kernel_.theta):.3f}')
```

```python
plt.figure(figsize=(20, 15))

X_ = np.linspace(X.min(), X.max() + 30, 1000)[:, np.newaxis]
y_pred, y_std = gp.predict(X_, return_std=True)

plt.scatter(X, y, c='k')
plt.plot(X_, y_pred)
plt.fill_between(X_[:, 0], y_pred - y_std, y_pred + y_std, alpha=0.5, color='k')
plt.xlim(X_.min(), X_.max())

plt.xlabel('Year')
plt.ylabel('CO2 in ppm')
plt.tight_layout()
plt.show()
```

```python
# Kernels with optimized parameters
k1 = 50.0 ** 2 * RBF(length_scale=50.0)
k2 = 2.0 ** 2 * RBF(length_scale=100.0) \
     * ExpSineSquared(length_scale=1.0, periodicity=1.0, periodicity_bounds='fixed')

k3 = 0.5 ** 2 * RationalQuadratic(length_scale=1.0, alpha=1.0)
k4 = 0.1 ** 2 * RBF(length_scale=0.1) \
     + WhiteKernel(noise_level=0.1 ** 2, noise_level_bounds=(1e-3, np.inf))

my_kernel = k1 + k2 + k3 + k4

gp = GaussianProcessRegressor(kernel=my_kernel, alpha=0, normalize_y=True)
gp.fit(X, y)

print(f'Kernel: {gp.kernel_}')
print(f'Log marginal likelihood: {gp.log_marginal_likelihood(gp.kernel_.theta):.3f}')
```

```python
plt.figure(figsize=(20, 15))

X_ = np.linspace(X.min(), X.max() + 30, 1000)[:, np.newaxis]
y_pred, y_std = gp.predict(X_, return_std=True)

plt.scatter(X, y, c='k')
plt.plot(X_, y_pred)
plt.fill_between(X_[:, 0], y_pred - y_std, y_pred + y_std, alpha=0.5, color='k')
plt.xlim(X_.min(), X_.max())

plt.xlabel('Year')
plt.ylabel('CO2 in ppm')
plt.tight_layout()
plt.show()
```
