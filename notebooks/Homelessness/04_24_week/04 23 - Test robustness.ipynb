```python
import pandas as pd
import numpy as np
np.random.seed(0)
from scipy.optimize import linear_sum_assignment

import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = 20, 13
plt.rcParams['font.size'] = 20
plt.rcParams['lines.linewidth'] = 2
import seaborn as sns

import sys
sys.path.insert(0, '../../..')
import leximin_assignment
import matrix_gen

from tqdm import tqdm
```

```python
def custom_eff_assignment(cost_matrix, capacities):
    full_cost_matrix = np.repeat(cost_matrix, capacities, axis=1)
    row_ind, col_ind = linear_sum_assignment(full_cost_matrix)
    
    return full_cost_matrix[row_ind, col_ind].sum()
```

```python
N = 30
N_INTVS = 5
N_EXPERIMENTS = 1000
CAPACITIES = np.ones((N_INTVS,), dtype=int) * (N // N_INTVS)
```

```python
matrix_generator = matrix_gen.MatrixGenerator(
    np.random.beta, (1e-2 * 3, 1e-1), N, N_INTVS
)
```

# Counting the number of 0-1 matrices

## Including 0-1 matrices

```python
count = 0
running_pof_sum = 0

for _ in range(N_EXPERIMENTS):
    (_, sorted_matrix), (__, sorted_increase_matrix) = matrix_generator.get_increase_matrix_pair(
        round_=False
    )
    
    if np.all(np.logical_or(np.equal(sorted_matrix, 0), np.equal(sorted_matrix, 1))):
        count += 1
        running_pof_sum += 1
        continue
    
#     print(sorted_matrix)
#     print(sorted_increase_matrix)
#     print()
    
    eff_cost = custom_eff_assignment(sorted_matrix, CAPACITIES)
    lex_assigner = leximin_assignment.RowSortedLeximinAssignmentHelper(
        sorted_increase_matrix, CAPACITIES
    )
    lex_assignments = lex_assigner.solve()
    lex_cost = lex_assigner.get_cost(
        lex_assignments, cost_matrix=sorted_matrix
    )
    
    if eff_cost != 0:
        running_pof_sum += lex_cost / eff_cost
    else:
        running_pof_sum += 1
```

```python
count, running_pof_sum / N_EXPERIMENTS
```

# Excluding 0-1 matrices

```python
success_count = 0
running_pof_sum = 0

while success_count < N_EXPERIMENTS:
    print(success_count)
    
    (_, sorted_matrix), (__, sorted_increase_matrix) = matrix_generator.get_increase_matrix_pair(
        round_=False
    )
    
    if np.all(np.logical_or(np.equal(sorted_matrix, 0), np.equal(sorted_matrix, 1))):
        # count += 1
        # running_pof_sum += 1
        continue
    
#     print(sorted_matrix)
#     print(sorted_increase_matrix)
#     print()
    
    eff_cost = custom_eff_assignment(sorted_matrix, CAPACITIES)
    lex_assigner = leximin_assignment.RowSortedLeximinAssignmentHelper(
        sorted_increase_matrix, CAPACITIES
    )
    lex_assignments = lex_assigner.solve()
    lex_cost = lex_assigner.get_cost(
        lex_assignments, cost_matrix=sorted_matrix
    )
    
    if eff_cost != 0:
        running_pof_sum += lex_cost / eff_cost
    else:
        running_pof_sum += 1
    success_count += 1
```
