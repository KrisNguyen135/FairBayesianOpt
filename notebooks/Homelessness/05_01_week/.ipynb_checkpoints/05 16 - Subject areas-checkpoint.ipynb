```python
groups = [
    'Algorithms', 'Applications', 'Data, Challenges, Implementations, and Software',
    'Deep Learning', 'Neuroscience and Cognitive Science', 'Optimization',
    'Probabilistic Methods', 'Reinforcement Learning and Planning',
    'Social Aspects of Machine Learning', 'Theory'
]
```

```python
df = pd.read_csv('../../../subject_areas.txt', header=None, sep='\t')
df
```

```python
subgroups = {group: [] for group in groups}

temp_group = ''

for item in df[0].to_list() + df[3].to_list():
    if not isinstance(item, str):
        continue
    
    topic = item.strip()
    
    if topic in groups:
        print(topic)
        temp_group = topic
    else:
        subgroups[temp_group].append(topic)
        
# for item in df[3].to_list():
#     if not isinstance(item, str):
#         continue
    
#     if item in groups:
#         print(item)
#         temp_group = item
#     else:
#         subgroups[temp_group].append(item.strip())    
    
# subgroups
```

```python
subgroups
```

```python
with open('neurips_20_subject_areas.md', 'w') as f:
    for group in subgroups:
        f.write(f'\n{group}:\n')
        for topic in subgroups[group]:
            f.write(f'- {topic}\n')
```
