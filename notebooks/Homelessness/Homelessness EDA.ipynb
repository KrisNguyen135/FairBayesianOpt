```python
timesort_df = df.sort_values(by='Week')
timesort_df.head()
```

```python
timesort_df['Week'].value_counts().plot.bar()
plt.show()
```

### Number of households entering each week

```python
week_groupby = timesort_df.groupby('Week')
count_series = week_groupby.count().iloc[:, 1]

plt.plot(count_series.index, count_series)
plt.show()
```

### Average reentry probability each week

```python
for type_ in types:
    mean_series = week_groupby.mean()[type_]
    
    plt.plot(mean_series.index, mean_series)
    plt.title(type_)
    plt.show()
```

### Total reentry probability each week

```python
for type_ in types:
    sum_series = week_groupby.sum()[type_]
    
    plt.plot(sum_series.index, sum_series)
    plt.title(type_)
    plt.show()
```
