```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
import matplotlib
matplotlib.rcParams.update({'font.size': 12})
import seaborn as sns

import sys
sys.path.insert(0, '../../..')
import kde_utils
```

```python
types = ['ES', 'PSH', 'TH', 'RRH', 'PREV']
df = pd.read_csv('../../../data/subset_data.csv', index_col=0)
```

```python
sns.pairplot(df[types])
plt.show()
```

# Bounded KDEs

```python
data = df['ES'].to_numpy()
data
```

```python
N = 1000

quartiles = np.linspace(0, 100, N)
quartile_x = np.percentile(data, quartiles)
quartile_x
```

```python
plt.hist(data, bins=100, density=True)
plt.show()
```

```python
transformed_data = np.log(data) - np.log(1 - data)
transformed_data
```

```python
trans_quartile_x = np.log(quartile_x) - np.log(1 - quartile_x)
trans_quartile_x
```

```python
plt.hist(transformed_data, bins=100, density=True)
plt.show()
```

## Sklearn

```python
from sklearn.neighbors import KernelDensity

kernel = 'gaussian'

bandwidth = 0.9 \
            * min(transformed_data.std(), np.subtract(*np.percentile(transformed_data, [75, 25])) / 1.34) \
            * transformed_data.size ** (-1 / 5)

kde = KernelDensity(kernel=kernel, bandwidth=bandwidth).fit(transformed_data.flatten()[:, np.newaxis])
kde
```

```python
kde.score(trans_quartile_x.reshape(-1, 1))
```

```python
log_p = kde.score_samples(trans_quartile_x.reshape(-1, 1))
pdf = np.exp(log_p)
pdf
```

```python
plt.hist(transformed_data, density=True, bins=50, alpha=0.2,
         label='True data')
plt.plot(trans_quartile_x, pdf,
         label='Gaussian KDE')

plt.legend()
plt.show()
```

```python
plt.hist(data, density=True, bins=50, alpha=0.2,
         label='True data')
plt.plot(quartile_x, pdf,
         label='Gaussian KDE')

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 15))

ax[0].hist(data, density=True, bins=50, alpha=0.2,
           color='blue', label='True data')

ax[0].hist(transformed_data, density=True, bins=50, alpha=0.2,
           color='red', label='Transformed data')
ax[0].plot(trans_quartile_x, pdf,
           label='KDE on transformed data', c='r')
ax[0].legend()

ax[1].plot(trans_quartile_x, pdf.cumsum() / pdf.sum(), c='r')

plt.show()
```

```python
# CDF on the transformed data: trans_quartile_x, pdf.cumsum() / pdf.sum()
plt.plot(quartile_x, pdf.cumsum() / pdf.sum())
plt.show()
```

```python
plt.hist(data, bins=50, alpha=0.2, density=True)
plt.plot(quartile_x, pdf)

plt.show()
```

## Linspace vs. quartiles

```python
x = np.linspace(0, 1, N + 2)[1: -1]  # trim the endpoints
transformed_x = np.log(x) - np.log(1 - x)
transformed_x
```

```python
log_p = kde.score_samples(transformed_x.reshape(-1, 1))
pdf = np.exp(log_p)
pdf
```

```python
plt.hist(transformed_data, density=True, bins=50, alpha=0.2,
         label='True data')
plt.plot(transformed_x, pdf,
         label='Gaussian KDE')

plt.legend()
plt.show()
```

```python
def get_cdf(point):
    trans_point = np.log(point) - np.log(1 - point)
    # p_trans_point = kde.evaluate(trans_point)
    
    return pdf[transformed_x < trans_point].sum() / pdf.sum()
```

```python
dx = x[1] - x[0]

plt.hist(data, density=True, bins=100, alpha=0.2,
         label='True data')

plt.plot(x, np.gradient(pdf.cumsum(), dx) / pdf.sum())

plt.show()
```

```python
plt.plot(x, [get_cdf(point) for point in x])
plt.show()
```

```python
cdf = np.array([get_cdf(point) for point in x])
dx = x[1] - x[0]

plt.hist(data, density=True, bins=100, alpha=0.2,
         label='True data')

plt.plot(x, np.gradient(cdf, dx))

plt.show()
```

# Sampler from transformed KDE

```python
data = df[types].to_numpy().flatten()
plt.hist(data, bins=50, density=True)
plt.show()
```

```python
kde = kde_utils.KDEHelper(np.log(data) - np.log(1 - data))
```

```python
samples = kde.sample_and_transform(size=(data.size, 1))

plt.hist(data, bins=50, density=True, alpha=0.2,
         label='Real data')
plt.hist(samples, bins=20, density=True, alpha=0.2,
         label='Data sampled from KDE')

plt.legend()
plt.show()
```
