```python
households = [i for i in range(1, df.shape[0])]
types = ['ES', 'PSH', 'TH', 'RRH', 'PREV']
```

```python
# To be 
def fairness(assignment_df):
    return np.max(df['Reentered'] - np.array([
        sum(assignment_df.loc[household, type_] * df.loc[household, type_] 
            for type_ in types) 
        for household in households
    ]))
```
